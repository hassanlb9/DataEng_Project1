```python
#we remove numbers in a col
df['text'] = df['text'].str.replace('\d+', '')
```

```python
#remove punctuation because they have not meaning in the model
df["text"] = df['text'].str.replace('[^\w\s]','')
```

```python
#to lower
df["text"] = df['text'].apply(lambda x:x.lower())
```

```python
#remove stop words
stops = set(stopwords.words("english"))
df["text"] = df['text'].apply(lambda x:' '.join([w for w in nltk.word_tokenize(x) if not w in stops]))
```

```python
df.head()
```

```python
#train the model
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
classifier = MultinomialNB()
x= df['selected_text']
y=df['sentiment']
```

```python
type(x)
```

```python
#transform text into vectors
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
vectorizer.fit(x)
vector = vectorizer.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(vector,y,test_size=0.3,random_state=42)
```

```python
model=classifier.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
print("accuracy_score:{}".format(accuracy_score(y_test,y_pred)))
```

```python
pickle.dump(vectorizer, file = open('vocab.pkl', 'wb'))
```

```python
def preprocess(text):
    text=[text]
    vector = vectorizer.transform(text)
    pred=model.predict(vector)
    

    return float(pred)
    
    
```
